### Detect Outliers using the IQR Method

```python
# Detect outliers in numerical variables using the IQR or Z-score method.

num_cols = df.select_dtypes(include=['int64',]).columns  # Select numerical columns

# IQR method to detect outliers
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # Lower 25% of data
    Q3 = data[column].quantile(0.75)  # Upper 75% of data
    IQR = Q3 - Q1                     # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR      # Lower limit
    upper_bound = Q3 + 1.5 * IQR      # Upper limit
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Check the number of outliers in each column
for col in num_cols:
    outliers = detect_outliers_iqr(df, col)
    print(f"Column {col} has {len(outliers)} outliers.")
```

## 5.3 How does the remote work period affect health indicators?

```python
# Ya≈ü gruplarƒ±nƒ± olu≈ütur
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60']
)

# Ya≈ü gruplarƒ±na g√∂re burnout oranlarƒ±nƒ± hesaplayalƒ±m
burnout_rates = (
    df.groupby(['Age_Group', 'Burnout_Level'])
    .size()
    .groupby(level=0)
    .apply(lambda x: 100 * x / x.sum())
    .unstack()
)

# Storytelling renk paleti
story_colors = {
    'High': '#E63946',   # kƒ±rmƒ±zƒ±
    'Medium': '#F4A261', # turuncu
    'Low': '#2A9D8F'     # ye≈üil
}

# G√∂rselle≈ütirme
burnout_rates.plot(
    kind='bar',
    figsize=(8,6),
    color=[story_colors.get(col, '#333333') for col in burnout_rates.columns]
)
plt.title("Ya≈ü Gruplarƒ±na G√∂re T√ºkenmi≈ülik (Burnout) Oranlarƒ± (%)", fontsize=14, fontweight='bold')
plt.ylabel("Oran (%)")
plt.xlabel("Ya≈ü Grubu")
plt.legend(title="Burnout D√ºzeyi")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

This visualization shows the distribution of burnout levels by age group.

The predominance of the 'Medium' level across all age groups indicates that employees generally experience a moderate level of stress.

However, the 'High' level rate is notably higher, especially in the 18‚Äì25 and 36‚Äì45 age groups.

This suggests that employees who are new to the workforce (young age) and those in the mid-career stage with increasing responsibilities (middle age) may be at a higher risk of burnout.

In summary: The highest stress level was observed in the 18‚Äì25 and 36‚Äì45 age groups.

### Targeting Burnout: Strategic Interventions

```python
# Create age groups
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 60], # We divided employees into 4 different age groups
    labels=['18-25', '26-35', '36-45', '46-60'] 
)

# Let's calculate burnout rates by age group
burnout_rates = (
    df.groupby(['Age_Group', 'Burnout_Level'])
    .size()
    .groupby(level=0)
    .apply(lambda x: 100 * x / x.sum())
    .unstack()
)

# Storytelling color palette
story_colors = {
    'High': '#E63946',   # red
    'Medium': '#F4A261', # orange
    'Low': '#2A9D8F'     # green
}

# Visualization
burnout_rates.plot(
    kind='bar',
    figsize=(8,6),
    color=[story_colors.get(col, '#333333') for col in burnout_rates.columns]
)
plt.title("Burnout Rates (%) by Age Group", fontsize=14, fontweight='bold')
plt.ylabel("Rate (%)")
plt.xlabel("Age Group")
plt.legend(title="Burnout Level")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Observation/Conclusion:

As shown in the graph above, the burnout rate is particularly higher among employees in the 26‚Äì45 age group.

This suggests that this age cohort may be at a higher risk due to increasing professional and personal responsibilities.

Recommendations (Targeted Interventions):

Therefore, specialized stress management workshops, coaching, or mentorship programs can be implemented specifically for this age group.

```python
#  Explode the data
# The 'Physical_Health_Issues' column often contains multiple issues separated by a semicolon.
# This assigns a new 'Issue' column by splitting the semicolon-separated strings and then 'exploding'
# the DataFrame to create a new row for each individual issue reported.
df_exploded = df.assign(Issue=df['Physical_Health_Issues'].str.split(';')).explode('Issue')
df_exploded['Issue'] = df_exploded['Issue'].str.strip()

#  Exclude 'None' entries
# Filters out rows where the reported issue is explicitly 'None' or case variations of 'none'.
df_exploded = df_exploded[df_exploded['Issue'].str.lower() != 'none']

# Reset the index
df_exploded = df_exploded.reset_index(drop=True)

#  Calculate the total number of issues reported per work arrangement
# This finds the total count of all reported issues within each Work_Arrangement group.
total_counts = df_exploded.groupby('Work_Arrangement')['Issue'].count().reset_index(name='Total')

# Count the frequency of each Issue - Work_Arrangement combination
issue_counts = (
    df_exploded.groupby(['Work_Arrangement', 'Issue'])
    .size()
    .reset_index(name='Count')
)

#  Calculate the percentage of each Issue within that specific work arrangement
# Merges the issue counts with the total counts to calculate the percentage of each issue relative
# to the total number of issues reported in that arrangement.
issue_counts = issue_counts.merge(total_counts, on='Work_Arrangement')
issue_counts['Percent'] = (issue_counts['Count'] / issue_counts['Total']) * 100

#  Plot the graph
plt.figure(figsize=(10,6))
sns.barplot(
    data=issue_counts,
    x='Issue',
    y='Percent',
    hue='Work_Arrangement',
    palette='Set2'
)
plt.title("Physical Health Issues by Work Arrangement (Percentage Based)", fontsize=14)
plt.xlabel("Type of Health Issue")
plt.ylabel("Rate (%)")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()

```
